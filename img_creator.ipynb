```python
import matplotlib.pyplot as plt

from file_utils import load_list_from_file
```

```python
errors1 = load_list_from_file("data/errors1.txt", float)
plt.plot(errors1)
plt.grid()
plt.ylabel('Error value 1')
plt.savefig("img/error1.png")
plt.show()
```

```python
its1 = load_list_from_file("data/gradient/its.txt", int)
its2 = load_list_from_file("data/gradient/opt_its.txt", int)

general_its = [(i1, i2) for i1, i2 in zip(its1, its2) if i1 != 0 and i2 != 0 and i1 > i2]
general_its = sorted(general_its, key=lambda item: item[1])
its1 = [item[0] for item in general_its]
its2 = [item[1] for item in general_its]

x = list(range(len(its1)))
bound = len(x)

plt.plot(x, its1[:bound], "r")
plt.plot(x, its2[:bound], "b")

plt.show()

plt.plot(x, sorted(its1)[:bound], "r")
plt.plot(x, sorted(its2)[:bound], "b")

plt.show()
```

```python
its1_1 = load_list_from_file("data/gradient/its.txt", int)
its1_2 = load_list_from_file("data/gradient/opt_its.txt", int)
its2_1 = load_list_from_file("data/conjugate/its.txt", int)
its2_2 = load_list_from_file("data/conjugate/opt_its.txt", int)

general_its_1 = [(i1, i2) for i1, i2 in zip(its1_1, its1_2) if i1 != 0 and i2 != 0 and i1 > i2]
general_its_2 = [(i1, i2) for i1, i2 in zip(its2_1, its2_2) if i1 != 0 and i2 != 0 and i1 > i2]
general_its_1 = sorted(general_its_1, key=lambda item: item[1])
general_its_2 = sorted(general_its_2, key=lambda item: item[1])

its1_1 = [item[0] for item in general_its_1]
its1_2 = [item[1] for item in general_its_1]
its2_1 = [item[0] for item in general_its_2]
its2_2 = [item[1] for item in general_its_2]

x = list(range(min(len(its1_1), len(its2_1))))
bound = len(x)

plt.plot(x, sorted(its1_1)[:bound], "r")
plt.plot(x, sorted(its1_2)[:bound], "b")
plt.plot(x, sorted(its2_1)[:bound], "g")
plt.plot(x, sorted(its2_2)[:bound], "m")

plt.show()
```
